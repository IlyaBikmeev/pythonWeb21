```python
#Факториал через цикл while

n = int(input())
factorial = 1
while n > 1:
    factorial *= n
    n -= 1
print(factorial)
```

```python
'''
Подается натуральное число N на вход. Определить, имеется ли в записи числа цифра 2

 1<= N <= 1000000000
 
 136275451
 
'''
n = int(input())
res = False
while n > 0:
    if n % 10 == 2:
        res = True
        break
  
    n //= 10 #n = n // 10
    
if res:
    print('Имеется')
else:
    print('Не имеется')
```

```python
'''
Подается на вход степень двойки. Нужно определить показатель этой степени
Примеры:
1 -> 0
2 -> 1
4 -> 2
8 -> 3

... 
1024 -> 10
2048 -> 11
'''

n = int(input())
k = 0

while n > 1:
    n /= 2 #Делим число на два n = n / 2
    k += 1 #Прибавляем 1 к счетчику
print(k)

#8 = 2 * 2 * 2 * 1
#1024 = 2 * 2 * 2 * 2 * 2 * 2 * 2 * 2 * 2 * 2 * 1
```

```python
x, eps = map(float, input().split())
prev_sum = 0
cur_sum = x
#abs - модуль числа. abs(-1) = 1. abs(2) = 2
k = 2 #Степень х и знаменатель

while abs(cur_sum - prev_sum) >= eps:
    prev_sum = cur_sum
    term = x ** k / k
    if k % 2 == 0:
        term *= -1
    cur_sum += term
    k += 1

print(cur_sum)
```

#### Посчитайте число Е с заданной точностью
![title](exponent.jpg)

```python
# Подается на вход точность eps. Вычислить число e с заданной точностью.
# Критерий остановки: посчитанное число отличается от встроенного math.e меньше eps
# 1 + 1/1! + 1/2! + 1/3! +1/4! +....
# Для факториала используйте math.factorial
```

```python
import math

eps = float(input())

s = 0
k = 0

while abs(s - math.e) >= eps:
    #print('Прибавляем 1/{}!'.format(k))
    s += 1 / math.factorial(k)
    k += 1
print(s)
```
