```python
#Циклы for

#Для каждого элемента из последовательности
for переменная in последовательность
```

```python
#1. Строка является последовательностью

s = 'строка'

#Для каждого символа из строки c
for c in s:
    print(c)
```

```python
#2. Генератор range() является последовательностью

#Пример: перебрать числа от 5 до 15

#Два параметра
for n in range(5, 16):
    print(n)

print()

#Один параметр
for n in range(16):
    print(n)
```

```python
#3 Список является последовательностью.

a = [1,2,3,4,5,5,4,3,2,1]

for n in a:
    print(n)
```

```python
# Пример с input().split() :
# '1 2 3 4 5 6 7 8'.split()  ['1', '2', '3', '4', '5', '6', '7', '8']

#Задача: посчитать сумму целых чисел, введенных через пробел

s = 0
for n in input().split():
    s += int(n)
print(s)
```

```python
#Задача: пользователь вводит какое-то количество целых чисел через пробел, выведите максимум из них

#Ввод: 1 75 1 346 8 2 4 7 246 4267 23 7 35 68
#Вывод: 4267
```

```python
ans = 0
for n in input().split():
    if int(n) > ans:
        ans = int(n)

print(ans)

```

```python
#Способ первый (не очень хороший)
ans = 0 #максимум
fl = True #Флаг для первой итерации

for n in input().split():
    if fl:
        ans = int(n)
        fl = False
        continue
    
    if int(n) > ans:
        ans = int(n)
    
print(ans)
        
```

```python
#Способ второй: считать изначально в массив
#list comprehension
a = [int(x) for x in input().split()]
ans = a[0]
for n in a:
    if n > ans:
        ans = n
print(ans)
```

```python
#Способ третий (в одну строчку)
print(max([int(x) for x in input().split()]))
```

```python
a = [int(x) for x in input().split()]

s = 0
M = max(a)
for i in range(100) :
    s += M
print(s)
```

```python
#Пользователь дает список чисел. Найти их сумму
#sum - принимает список, выдает сумму чисел в нем
print(sum([int(x) for x in input().split()]))
```

```python
#Функция sum реализована следующим образом:
a = [int(x) for x in input().split()]
s = 0
for n in a:
    s += n
print(s)

```

```python
s = 0
n = int(input())
for k in range(1, n + 1):
    s += k ** k
print(s)
```

```python
#Подается список чисел, нужно найти среднее арифметическое всех чисел, которые делятся на 3

s = 0   #Сумма
cnt = 0    #количество
for n in input().split():
    n = int(n)
    if n % 3 == 0:
        cnt += 1
        s += n
print(s / cnt)
```
