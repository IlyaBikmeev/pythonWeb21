```python
#У строки есть метод split
s = "1 2 3 4 5"
l = s.split()

#А как наоборот? Список в строку (метод join)

#Способ 1
l = [1, 2, 3, 4, 5]
for i in range(len(l)):
    l[i] = str(l[i]) #преобразовываем i-й элемент в строку
    
s = "разделитель".join(l)
print(s)
```

```python
#List comprehension
#Мы можем создать список циклом с каким-то условием
#a = [выражение(элемент) for элемент in последовательность if условие]
```

```python
#Пример:
a = [1, 63, 13, 754, 13, 264, 416, 575]
#хотим создать новый список, в котором будут четные элементы из a
#Как мы делали бы это до этого занятия:
ans = []
for x in a:
    if x % 2 == 0:
        ans.append(x)
print(ans)

#Как мы бы делали через list comprehension

ans_2 = [x for x in a if x % 2 == 0]
print(ans_2)
```

```python
#Подается натуральное число n. Создать список из квадратов 
#натуральных чисел от 1 до n включительно
#Пример: n = 5, список: [1, 4, 9, 16, 25]

n = int(input())
a = []
for i in range(1, n + 1):
    a.append(i * i)
print(a)

#Через list comprehension
a = [x * x for x in range(1, n + 1)]
print(a)
```

```python
#Вводится число n. Получить массив, в котором i-й элемент равен 5*i+7
# i = 1, n
#Пример: ввод n = 3
#Вывод : [12, 17, 22]

n = int(input())
a = [5 * i + 7 for i in range(1, n + 1)]
print(a)
```

```python
#Вводятся числа N, M. Создать массив размера N * M 
#и заполнить его случайными целыми числами от 1 до 100 
import random

N, M = map(int, input().split())

A = []

for i in range(N):
    row = []
    for j in range(M):
        row.append(int(random.randrange(1, 100)))
    A.append(row)

    
for row in A:
    for col in row:
        print(col, end=' ')
    print()
```

```python
#Вводится число N. Затем вводится матрица N*N. Выведите сумму по строкам
#Пример: 
#Ввод:
#    3
#    1 2 3
#    4 5 6
#    7 8 9
#Вывод:
#    6
#    15
#    24
'''
Считывание 
N = int(input())
A = [[int(x) for x in input().split()] for i in range(N)]
'''
```

```python
N = int(input())
A = [[int(x) for x in input().split()] for i in range(N)]

for i in range(N):
    s = 0
    for j in range(N):
        s += A[i][j]
    print(s)
```
