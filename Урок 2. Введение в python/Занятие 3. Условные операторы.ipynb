```python
'''
Электронные часы показывают время в формате h:mm:ss (от 0:00:00 до 23:59:59), 
то есть сначала записывается количество часов, потом обязательно двузначное количество минут, 
затем обязательно двузначное количество секунд. Количество минут и секунд при необходимости 
дополняются до двузначного числа нулями.

С начала суток прошло N секунд. Выведите, что покажут часы.
На вход дается натуральное число N, не превосходящее 10000000
'''
totalSeconds = int(input())

hours = (totalSeconds // 3600) % 24 #Получаем сколько часов в данном количестве секунд
minutes = (totalSeconds % 3600) // 60 #Получаем сколько минут
seconds = (totalSeconds % 60)

#time- переменная-строка для результата
time = str(hours) + ':' + str(minutes // 10) + str(minutes % 10) + ':' + str(seconds // 10) + str(seconds % 10) 
print(time)
```

```python
'''
Следующее четное
На вход дается натуральное число N. Выведите следующее за ним четное число

5 -> 6
6 -> 8

((n // 2) + 1 ) * 2

n + 2 - n % 2 = (n + 2) - (n - (n // 2) * 2 ) = n + 2 + n + (n // 2) * 2 = 2 * (n // 2 + 1)

'''

n = int(input())
print(n + 2 - n % 2)


```

```python
'''
МКАД

Длина Московской кольцевой автомобильной дороги — 109 километров. 
Байкер Вася стартует с нулевого километра МКАД и едет со скоростью V километров в час. 
На какой отметке он остановится через T часов?

Формат входных данных

Программа получает на вход целые числа V и T. Если V > 0, 
то Вася движется в положительном направлении по МКАД, если же значение V < 0,
то в отрицательном. 0 ≤ T ≤ 1000, -1000 ≤ V ≤ 1000
'''

v = int(input()) #Скорость
t = int(input()) #Время

position = ((v * t + 109 * t) + 109) % 109 #Если нужно убрать отрицательность в остатке

print(position)
```
