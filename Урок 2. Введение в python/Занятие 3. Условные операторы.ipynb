#### Задание
Даны три стороны треугольника a, b, c. Выведите площадь такого треугольника. Гарантируется, что треугольник существует
![jupyter](./img/geron.png)

##### Формат ввода:
Вводятся три вещественных числа a, b, c.

##### Формат вывода:
Вывести площадь треугольника со сторонами a, b, c.


```python
#Решение
import math

a = float(input())
b = float(input())
c = float(input())

p = (a + b + c) / 2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))
print(S)

```

#### Задание 
Написать программу вычисления **y** по формуле:

![jupyter](./img/formula.gif)

##### Ввод
> В первой строке содержится вещественное число x.

##### Вывод
> Вывести значение y с шестью десятичными знаками после запятой. Вывести "ERROR", если функция не может быть вычислена в области вещественных чисел.



```python
import math
#and - и, or - или

x = float(input())
y = 0
res = True #Отвечает за то, что функция может быть вычислена
if x < -0.5:
    if math.cos(x) >= 0:
        y = 1 + math.sqrt(math.cos(x)) 
    else:
        res = False
    
elif x >= -0.5 and x <= 1:
    y = x + 1
else:
    y = 1 - x ** 2
    
#Вывод с 6 знаками после запятой

if res:
    print("{:.6}".format(y))
else:
    print("ERROR")
```

```python
#Задание 3
import math

#a, b, c = *[float(x) for x in input().split()]

a = float(input())
b = float(input())
c = float(input())

d = b ** 2 - 4 * a * c

if d < 0:
    print("Нет корней")
elif d == 0:
    x = -b / (2 * a)
    print('%.2f' % x)
else:
    x_1 = (-b + math.sqrt(d)) / (2 * a)
    x_2 = (-b - math.sqrt(d)) / (2 * a)
    print('%.2f' % x_1, '%.2f' % x_2)

```

```python
#Задача 5

#Окружность
x_c = float(input())
y_c = float(input())
r = float(input())

#Точка
x_0 = float(input())
y_0 = float(input())

if (x_0 - x_c) ** 2 + (y_0 - y_c) ** 2 <= r ** 2:
    print('Точка внутри окружности')
else:
    print('Точка вне окружности')
```

```python
#Реализация, чтобы можно было вводить в одну строку через пробелы

#В одной строке окружность

x_c, y_c, r = [float(x) for x in input().split()]
x_0, y_0 = [float(x) for x in input().split()]
if (x_0 - x_c) ** 2 + (y_0 - y_c) ** 2 <= r ** 2:
    print('Точка внутри окружности')
else:
    print('Точка вне окружности')
```
